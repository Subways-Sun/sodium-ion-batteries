```python
import matplotlib.pyplot as plt

# Sample data
categories = ["Pre-2000", "2000-2004", "2005-2009", "2010-2014"] + [str(year) for year in range(2015, 2025)]
values = [89, 112, 171, 1129, 874, 1168, 1618, 2167, 2373, 2521, 2912, 2901, 2910, 3910]

# Create the bar chart
plt.figure(figsize=(18, 5))
bars = plt.bar(categories, values)
for bar in bars:
    # Get the height (value) of the bar
    height = bar.get_height()
    # Place text with slight vertical offset above the bar
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x position
        height + 0.5,                      # y position
        str(height),                       # text (the value)
        ha='center',                       # horizontal alignment
        va='bottom'                        # vertical alignment
    )

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Amount of Papers on SIB each year")

# Display the plot
plt.show()
```
